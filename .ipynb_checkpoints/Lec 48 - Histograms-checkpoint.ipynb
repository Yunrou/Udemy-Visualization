```python
# Let's begin by importing what we'll need (You'll probably be copying and pasting this a lot)

# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats librayr from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

First of all, source of information for what a histogram actually is:
http://en.wikipedia.org/wiki/Histogram

```python
#Create a random normal-dist dataset
dataset1 = randn(100)

#Plot a histogram of the dataset, note bins=10 by default
plt.hist(dataset1)
```

```python
# Lets make another dataset
dataset2 = randn(80)

#Plot
plt.hist(dataset2,color='indianred')
```

```python
# We can use normed to plot on same plot

# Set normed=True for the plots to be normalized in order to comapre data sets with different number of observations
# Set alpha=0.5 for transperancy

plt.hist(dataset1,normed=True,color='indianred',alpha=0.5,bins=20)
plt.hist(dataset2,normed=True,alpha=0.5,bins=20)
```

```python
# Make two more random normal dist data sets
data1 = randn(1000)
data2 = randn(1000)

#Can represent joint distributions using joint plots
sns.jointplot(data1,data2)
```

```python
# Can also use hex bins for a more concise picture
sns.jointplot(data1,data2,kind='hex')
```

```python
# Standard
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
# Stats
from scipy import stats
```

```python
# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
plt.style.use('seaborn') # ggplot
sns.set_style("darkgrid")
```

```python
url = 'http://en.wikipedia.org/wiki/Histogram'
```

```python
dataset1 = randn(100)
```

```python
# separate your dataset into 10 equally-spaced bins
plt.hist(dataset1, edgecolor='white')
```

```python
dataset2 = randn(8)
plt.hist(dataset2, color='indianred', edgecolor='white')
```

```python
# normed -> density
plt.hist(dataset1, density=True, color='indianred', alpha=0.5, bins=20, edgecolor='white')
plt.hist(dataset2, density=True, alpha=0.5, bins=20, edgecolor='white')
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
# use `arr[tuple(seq)]` instead of `arr[seq]`.
sns.jointplot(tuple(data1), tuple(data2))
```

```python
sns.jointplot(tuple(data1), tuple(data2), kind='hex')
```
